#### 2.1. Изучите распределение домов от наличия вида на набережную
Постройте график\
Сделайте выводы

```python
data = df_house ['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

**Вывод:** из общей массы, домов с видом на набережную всего 0,8 %.

#### 2.2. Изучите распределение этажей домов

```python
data = df_house ['floors'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

**Вывод:** больше всего одноэтажных домов.

#### 2.3. Изучите распределение состояния домов

```python
data = df_house['condition'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

**Вывод:** больше всего домов, находящих в среднем состояни - 64,9 %.

## **Условие 3:**
### *3 задача*
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df_house.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df_house['price'], y=df_house['sqft_living'], kind='reg');
```

**Вывод:** Чем больше жилплощадь дома, тем выше его стоимость.

```python
sns.jointplot(x=df_house ['price'], y=df_house['grade'], kind='reg');
```

**Вывод:** чем выше качество конструкции и дизайна дома, тем выше его стоимость.

```python
sns.boxplot(x=df_house['price'], y=df_house['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

**Вывод:** Наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13.

```python
sns.boxplot(x=df_house['price'], y=df_house['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

**Вывод:** Наибольшую стоимость имеют дома с количеством этажей 2,5.

```python
sns.boxplot(x=df_house['price'], y=df_house['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

**Вывод:** Наибольшую стоимость имеют дома с оценкой вида 4.
